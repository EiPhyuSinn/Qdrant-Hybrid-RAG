```python
import json 

with open('documents.json','r') as f:
    docs = json.load(f)

```

```python
documents_flatten = []

for doc in docs:
    for q_a in doc.get('documents'):
        q_a['course'] = doc.get('course')
        documents_flatten.append(q_a)

len(documents_flatten)
```

```python
import minsearch
Index = minsearch.Index(
    text_fields = ['text','section','question'],
    keyword_fields = ['course'])
```

```python
import minsearch

Index = minsearch.Index(
    text_fields = ['text','section','question'],
    keyword_fields = ['course'])
Index.fit(documents_flatten)
    
```

```python
q = 'When will the course start?'
```

```python
results = Index.search(q,num_results=5)

for r in results:
    print('\n',r['course'])
    print(r['text'])
```

```python
results = Index.search(q, filter_dict = {'course': 'data-engineering-zoomcamp'},num_results=3)
for r in results:
    print('\n',r['course'])
    print(r['text']) 
```

```python
boost_dict = {'text':2.0,'section':0.5,'question':2.0}
results = Index.search(q, filter_dict = {'course': 'data-engineering-zoomcamp'},boost_dict=boost_dict,num_results=3)
for r in results:
    print(r.keys())
```
